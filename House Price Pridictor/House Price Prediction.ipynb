```python
X=data[['number of bedrooms','number of bathrooms','living area','condition of the house','Number of schools nearby']]
y=data[['Price']]
```

```python
X,y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.tree import DecisionTreeRegressor
param_grid = {
    "criterion" : ["squared_error","friedman_mse","absolute_error"],
    "splitter" : ["best","random"],
    "max_depth" : [None,10,20,30,40,50],
    "min_samples_split" : [2,5,10],
    "min_samples_leaf" : [1,2,3]
}
```

```python
tree_model=DecisionTreeRegressor()
```

```python
grid_tree=GridSearchCV(estimator=tree_model,param_grid=param_grid,cv=5)
```

```python
grid_tree.fit(X_train,y_train)
```

```python
grid_tree.best_params_
```

```python
tree_preds=grid_tree.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,tree_preds)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
predslr=lr.predict(X_test)
```

```python
mean_squared_error(y_test,predslr)
```

```python
from sklearn.metrics import mean_absolute_error
mean_squared_error(y_test,tree_preds)
```

```python
from sklearn.metrics import mean_absolute_error
mean_squared_error(y_test,predslr)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfmodel=RandomForestRegressor()
```

```python
param_grid_fr={
    "max_depth" : [5,10,15],
    "n_estimators" : [2,3,4,5,6,7,8,9,10]
}
```

```python
gridfr=GridSearchCV(rfmodel,param_grid_fr)
```

```python
gridfr.fit(X_train, y_train.values.ravel())
```

```python
gridfr.best_params_
```

```python
rfpredictions=gridfr.predict(X_test)
```

```python
mean_absolute_error(y_test,rfpredictions)
```
